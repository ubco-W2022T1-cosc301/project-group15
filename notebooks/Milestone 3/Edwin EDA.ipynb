```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

dataset = pd.read_csv("../../data/raw/all_games.csv")

dataset
```

```python
# number of duplicate names
print(18800 - len(dataset['name'].unique()))
```

```python
dataset['platform'].value_counts()
```

```python
sns.countplot(y='platform', data=dataset, order=dataset['platform'].value_counts().index)
```

### Note:
We seem to have a lot of PC games in this dataset

```python
platforms = dataset['platform'].unique()

for platform in platforms:
    print(f"{platform} | {dataset[dataset['platform'] == platform]['meta_score'].mean()}")

sns.barplot(y='platform', x='meta_score', data=dataset)
```

### Note:
The game ratings between device types are fairly similar
